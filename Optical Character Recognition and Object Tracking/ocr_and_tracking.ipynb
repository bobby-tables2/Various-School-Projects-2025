# Translate text in image
Capture text (must be in the English alphabet) and translate it to a target language

```python
from PIL import Image
import pytesseract
from deep_translator import GoogleTranslator

# Simple image to string
image_path = input("Enter path to image")
language = input("Enter language to translate to")

langs_dict = GoogleTranslator().get_supported_languages(as_dict=True)
language = langs_dict[language]

extracted_text = pytesseract.image_to_string(Image.open(image_path))
translated = GoogleTranslator(source='auto', target=language).translate(extracted_text)
print(translated)
```

# Track objects in video
Either categorise all the objects in the video or track a specfifc one by their id.

```python
from ultralytics import YOLO
import cv2

# load model
model = YOLO('yolov8n.pt')

# load video
video_path = input("Enter path to video")
cap = cv2.VideoCapture(video_path)

selected_id = -1

frame = 1
while frame is not None:
    _, frame = cap.read()

    # detect and track objects
    results = model.track(frame, persist=True)

    frame_ = None

    # plot all results
    if(selected_id == -1):
        frame_ = results[0].plot()
    else:
        # plot result associated with specific id
        for result in results[0]:
            # Filter results for the vehicle of interest
            if result.boxes.id == selected_id:
                frame_ = result.plot()
                break

    # visualise
    try:
        cv2.imshow('frame', frame_)
    except:
        # could not find object
        selected_id = -1

    if cv2.waitKey(10) & 0xFF == ord('q'):
        selected_id = input("Type id to focus on, type -1 to show all, type q to break")

        if selected_id == 'q':
            cv2.destroyAllWindows()
            break
        else:
            selected_id = int(selected_id)

cv2.destroyAllWindows()
```
